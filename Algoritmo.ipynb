```python
import getpass

# Diccionarios para almacenar datos
citas = {
    "123456789": {
        "dÃ­a": "01/12/2024",
        "hora": "10:30",
        "tipo": "General",
        "mÃ©dico": "Dr. PÃ©rez",
        "estado": "Pendiente"
    },
    "987654321": {
        "dÃ­a": "02/12/2024",
        "hora": "14:00",
        "tipo": "Especialista",
        "mÃ©dico": "Dra. RamÃ­rez",
        "estado": "Confirmada"
    }
}

atenciones = [
    "Cita agregada para cÃ©dula 123456789.",
    "Cita confirmada para cÃ©dula 987654321."
]

# ContraseÃ±a del administrador
ADMIN_PASS = "rana123"

# Dibujos de ranas en ASCII
RANA_BIENVENIDA = r"""
        @..@
       (----)
      ( >__< )
      ^^ ~~ ^^
  Sana, sana, colita de rana
"""

RANA_DESPEDIDA = r"""
        @..@
       (----)
      ( >__< )
      ^^ ~~ ^^
 Â¡Que tengas un dÃ­a saltarÃ­n y feliz!
"""

RANA_MENU = r"""
        @..@
       (----)
      ( >__< )
      ^^ ~~ ^^
"""

# Funciones auxiliares para diseÃ±o
def imprimir_encabezado(titulo):
    #Imprime un encabezado con estilo y temÃ¡tica de ranas.
    print("\n" + "=" * 50)
    print(f"ðŸ¸ {titulo.center(46)} ðŸ¸")
    print("=" * 50)

def imprimir_opciones(opciones):
    #Imprime un menÃº de opciones con estilo.
    for num, opcion in enumerate(opciones, 1):
        print(f"  {num}. {opcion}")

def imprimir_separador():
    #Imprime un separador visual.
    print("-" * 50)

def imprimir_rana_menu():
    #rana en ASCII para embellecer los menÃºs.
    print(RANA_MENU)

######################################################################################################

#Funciones del programa

######################################################################################################
def agregar_cita():
    imprimir_encabezado("Agregar Cita ðŸ¸")
    imprimir_rana_menu()
    cedula = input("Ingrese la cÃ©dula del paciente: ")
    if cedula in citas:
        print("âš ï¸ Ya existe una cita para esta cÃ©dula.")
        return

    # Recolectar datos de la cita
    dia = input("Ingrese el dÃ­a de la cita (dd/mm/aaaa): ")
    hora = input("Ingrese la hora de la cita (hh:mm): ")
    tipo = input("Ingrese el tipo de cita (General/Especialista): ")
    medico = input("Ingrese el nombre del mÃ©dico: ")

    # Registrar cita
    citas[cedula] = {
        "dÃ­a": dia,
        "hora": hora,
        "tipo": tipo,
        "mÃ©dico": medico,
        "estado": "Pendiente"
    }
    atenciones.append(f"Cita agregada para cÃ©dula {cedula}.")
    print("âœ… Cita registrada exitosamente. Â¡La rana estÃ¡ feliz!")

def consultar_cita():
    imprimir_encabezado("Consultar Cita ðŸ¸")
    imprimir_rana_menu()
    cedula = input("Ingrese su cÃ©dula: ")
    if cedula in citas:
        print("\nDatos de su cita:")
        imprimir_separador()
        for clave, valor in citas[cedula].items():
            print(f"  {clave.capitalize()}: {valor}")
        imprimir_separador()
        print("ðŸ¸ Â¡Que tengas un dÃ­a feliz como una rana!")
    else:
        print("âš ï¸ No se encontrÃ³ ninguna cita asociada a esta cÃ©dula.")

def confirmar_cancelar_cita():
    imprimir_encabezado("Confirmar/Cancelar Cita ðŸ¸")
    imprimir_rana_menu()
    cedula = input("Ingrese la cÃ©dula del paciente: ")
    if cedula not in citas:
        print("âš ï¸ No se encontrÃ³ ninguna cita asociada a esta cÃ©dula.")
        return

    imprimir_opciones(["Confirmar cita", "Cancelar cita"])
    opcion = input("Seleccione una opciÃ³n: ")
    if opcion == "1":
        citas[cedula]["estado"] = "Confirmada"
        atenciones.append(f"Cita confirmada para cÃ©dula {cedula}.")
        print("âœ… La cita ha sido confirmada. Â¡La rana dice croac de felicidad!")
    elif opcion == "2":
        citas[cedula]["estado"] = "Cancelada"
        atenciones.append(f"Cita cancelada para cÃ©dula {cedula}.")
        print("âœ… La cita ha sido cancelada. Â¡La rana espera verte pronto!")
    else:
        print("âš ï¸ OpciÃ³n invÃ¡lida.")

def imprimir_reporte():
    imprimir_encabezado("Reporte de Citas ðŸ¸")
    imprimir_rana_menu()
    if not citas:
        print("âš ï¸ No hay citas registradas.")
        return
    for cedula, datos in citas.items():
        print(f"\nðŸ“„ CÃ©dula: {cedula}")
        imprimir_separador()
        for clave, valor in datos.items():
            print(f"  {clave.capitalize()}: {valor}")
        imprimir_separador()

    print("\n--- Historial de Atenciones ---")
    imprimir_separador()
    for registro in atenciones:
        print(f"  {registro}")
    print("ðŸ¸ Â¡Croac! Reporte finalizado.")

def menu_administrador():
    imprimir_encabezado("MenÃº Administrador ðŸ¸")
    password = getpass.getpass("Ingrese la contraseÃ±a del administrador: ")
    if password != ADMIN_PASS:
        print("âŒ ContraseÃ±a incorrecta. ðŸ¸")
        return

    while True:
        imprimir_encabezado("Opciones de Administrador ðŸ¸")
        imprimir_opciones(["Agregar cita", "Confirmar/Cancelar cita", "Imprimir reporte", "Salir al menÃº principal"])
        opcion = input("Seleccione una opciÃ³n: ")

        if opcion == "1":
            agregar_cita()
        elif opcion == "2":
            confirmar_cancelar_cita()
        elif opcion == "3":
            imprimir_reporte()
        elif opcion == "4":
            print("ðŸ¸ Saliendo al menÃº principal... Â¡Hasta luego, administrador rana!")
            break
        else:
            print("âš ï¸ OpciÃ³n invÃ¡lida.")

def menu_principal():
    while True:
        imprimir_encabezado("MenÃº Principal ðŸ¸")
        imprimir_rana_menu()
        imprimir_opciones(["Consultar cita", "MenÃº administrador", "Salir"])
        opcion = input("Seleccione una opciÃ³n: ")


        if opcion == "1":
            consultar_cita()
        elif opcion == "2":
            menu_administrador()
        elif opcion == "3":
            print(RANA_DESPEDIDA)
            break
        else:
            print("âš ï¸ OpciÃ³n invÃ¡lida.")

# EjecuciÃ³n del programa
if __name__ == "__main__":
    print(RANA_BIENVENIDA)
    print("Â¡Bienvenido a Sana, sana, colita de rana!")
    menu_principal()
```
